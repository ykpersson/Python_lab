### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

The two alternatives are the object-oriented and the Pyplot approach, the OO approach is recommended.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: is the whole plotted area, the background rectangle, text, axis, and the data.                                  Axis is the X, Y, Z axies, the coordinates. Artist are items plotted for example, chart titles, figure legends, colorbars, etc.), and subfigures.

### When plotting in Matplotlib, what is the expected input data type?

Numpy Arrays

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-4,4)
y = x**2
plt.plot(x, y)
plt.title("Grafen f(x)= X^2, Pyplot version")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-4,4)
y = x**2
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1]) 
axes.plot(x, y) 
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")
axes.set_title("grafen av f(x)= x^2, med metod objektorienterat API ")
axes.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,7))
fig.suptitle("Two charts, a scatterplot of X^2 and a barchart illustrating fruit availability")
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2
ax1.scatter(x, y)
ax1.set(xlabel='x-label', ylabel='y-label')
# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
ax2.bar(names, values)
plt.xlabel("Various fruits")
plt.ylabel("Amount")
plt.show()
```
